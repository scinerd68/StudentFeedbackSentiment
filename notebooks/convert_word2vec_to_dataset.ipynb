```python
from datasets import load_dataset
dataset = load_dataset("text", data_files="/content/word2vec_vi_words_300dims.txt")
```

```python
def convert_to_vector(line):
    line['text'] = line['text'].replace(u'\xa0', u'_')
    word, *embed = line['text'].split()
    embed = [float(num) for num in embed]
    line['word'] = word
    line['embed'] = embed
    return line
```

```python
dataset = dataset.map(convert_to_vector, remove_columns='text')
```

```python
dataset = dataset['train']
print(dataset)
```
